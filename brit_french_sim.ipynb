```python
# Function to compute sentence embeddings for an entire text
def compute_sentence_embeddings(text):
    sentences = tokenize_text(text)
    # Use the SentenceTransformer model to encode each sentence
    embeddings = model.encode(sentences, convert_to_tensor=True)
    return embeddings

# Compute embeddings for all the texts
garnier_embeddings = compute_sentence_embeddings(garnier_text)
mill_embeddings = compute_sentence_embeddings(mill_text)
say_embeddings = compute_sentence_embeddings(say_text)
frank_embeddings = compute_sentence_embeddings(frank_text)
blanqui_embeddings = compute_sentence_embeddings(blanqui_text)
dunoyer_embeddings = compute_sentence_embeddings(dunoyer_text)
malthus_embeddings = compute_sentence_embeddings(malthus_text)
mis_embeddings = compute_sentence_embeddings(mis_text)
smith_embeddings = compute_sentence_embeddings(smith_text)
ricardo_embeddings = compute_sentence_embeddings(ricardo_text)
rau_embeddings = compute_sentence_embeddings(rau_text)
rocher_embeddings = compute_sentence_embeddings(rocher_text)

# Check the shape of the embeddings
print("Shape of Garnier embeddings:", garnier_embeddings.shape)
print("Shape of Mill embeddings:", mill_embeddings.shape)
print("Shape of Say embeddings:", say_embeddings.shape)
print("Shape of Frank embeddings:", frank_embeddings.shape)
print("Shape of Blanqui embeddings:", blanqui_embeddings.shape)
print("Shape of Dunoyer embeddings:", dunoyer_embeddings.shape)
print("Shape of Malthus embeddings:", malthus_embeddings.shape)
print("Shape of Mis embeddings:", mis_embeddings.shape)
print("Shape of Smith embeddings:", smith_embeddings.shape)
print("Shape of Ricardo embeddings:", ricardo_embeddings.shape)
print("Shape of Rau embeddings:", rau_embeddings.shape)
print("Shape of Rocher embeddings:", rocher_embeddings.shape)


```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Function to compute the semantic similarity between two aggregated embeddings
def compute_similarity(embedding1, embedding2):
    # Move tensors to CPU and convert to NumPy arrays
    embedding1 = embedding1.cpu().numpy()
    embedding2 = embedding2.cpu().numpy()

    similarity = cosine_similarity([embedding1], [embedding2])[0][0]
    return similarity

# Calculate the average embedding for each text
garnier_aggregated = garnier_embeddings.mean(axis=0)
mill_aggregated = mill_embeddings.mean(axis=0)
say_aggregated = say_embeddings.mean(axis=0)
frank_aggregated = frank_embeddings.mean(axis=0)
blanqui_aggregated = blanqui_embeddings.mean(axis=0)
dunoyer_aggregated = dunoyer_embeddings.mean(axis=0)
malthus_aggregated = malthus_embeddings.mean(axis=0)
mis_aggregated = mis_embeddings.mean(axis=0)
smith_aggregated = smith_embeddings.mean(axis=0)
ricardo_aggregated = ricardo_embeddings.mean(axis=0)
rau_aggregated = rau_embeddings.mean(axis=0)
rocher_aggregated = rocher_embeddings.mean(axis=0)

# Calculate and print the pairwise similarities
texts = [
    (garnier_aggregated, "Garnier"),
    (mill_aggregated, "Mill"),
    (say_aggregated, "Say"),
    (frank_aggregated, "Frank"),
    (blanqui_aggregated, "Blanqui"),
    (dunoyer_aggregated, "Dunoyer"),
    (malthus_aggregated, "Malthus"),
    (mis_aggregated, "Mis"),
    (smith_aggregated, "Smith"),
    (ricardo_aggregated, "Ricardo"),
    (rau_aggregated, "Rau"),
    (rocher_aggregated, "Rocher")
]

# Loop through the texts and compute the pairwise similarities
for i in range(len(texts)):
    for j in range(i + 1, len(texts)):
        embedding1, name1 = texts[i]
        embedding2, name2 = texts[j]
        similarity = compute_similarity(embedding1, embedding2)
        print(f"Similarity between {name1} and {name2}: {similarity:.4f}")
```

```python
# Separate authors by nationality
# Add nationality information to the 'texts' list
texts = [
    (garnier_aggregated, "Garnier", "French"),
    (mill_aggregated, "Mill", "British"),
    (say_aggregated, "Say", "French"),
    (frank_aggregated, "Frank", "Fiction"),
    (blanqui_aggregated, "Blanqui", "French"),
    (dunoyer_aggregated, "Dunoyer", "French"),
    (malthus_aggregated, "Malthus", "British"),
    (mis_aggregated, "Mis", "Fiction"),
    (smith_aggregated, "Smith", "British"),
    (ricardo_aggregated, "Ricardo", "British"),
    (rau_aggregated, "Rau", "German"),
    (rocher_aggregated, "Rocher", "German")
]

french_authors = [(emb, name) for emb, name, nationality in texts if nationality == "French"]
british_authors = [(emb, name) for emb, name, nationality in texts if nationality == "British"]
german_authors = [(emb, name) for emb, name, nationality in texts if nationality == "German"]
fiction_authors = [(emb, name) for emb, name, nationality in texts if nationality == "Fiction"]

# Calculate average embeddings
french_avg_embedding = sum(emb for emb, _ in french_authors) / len(french_authors)
british_avg_embedding = sum(emb for emb, _ in british_authors) / len(british_authors)
german_avg_embedding = sum(emb for emb, _ in german_authors) / len(german_authors) #Use german_authors instead of french_authors
fiction_avg_embedding = sum(emb for emb, _ in fiction_authors) / len(fiction_authors) #Use fiction_authors instead of british_authors

# Compare French and British
french_british_similarity = compute_similarity(french_avg_embedding, british_avg_embedding)
print(f"Similarity between French and British authors: {french_british_similarity:.4f}")

# Define other authors
other_authors = german_authors + fiction_authors #Combine German and Fiction authors into other_authors

# Compare each group to Frank, Mis, Rau, and Rocher (if applicable)
for other_emb, other_name in other_authors:
    french_other_similarity = compute_similarity(french_avg_embedding, other_emb)
    british_other_similarity = compute_similarity(british_avg_embedding, other_emb)
    print(f"Similarity between French and {other_name}: {french_other_similarity:.4f}")
    print(f"Similarity between British and {other_name}: {british_other_similarity:.4f}")
```

```python
from sklearn.model_selection import train_test_split

# Split the data (you can adjust test_size as needed)
X_train, X_test, y_train, y_test = train_test_split(
    all_embeddings, labels, test_size=0.2, random_state=42, stratify=labels
)
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode the labels to integers
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

# Get the mapping from classes to integer labels
class_names = label_encoder.classes_
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize the classifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the classifier on the training data
clf.fit(X_train, y_train_encoded)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Predict on the test set
y_pred = clf.predict(X_test)

# Generate classification report
print("Classification Report:")
print(classification_report(y_test_encoded, y_pred, target_names=class_names))

# Confusion matrix
cm = confusion_matrix(y_test_encoded, y_pred)
print("Confusion Matrix:")
print(cm)
```
